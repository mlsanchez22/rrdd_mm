```r
read_files <- function(x){
    file <- paste("/mnt/lustre/scratch/CBRA/research/projects/rrdd_mm/results/05_clasi_result_kstest/05_2_clasi_result_kstest_sign_only_female/", x,sep="")
    kstest_pv_adj_kovsctrl_sign <- readRDS(file)
    kstest_pv_adj_kovsctrl_sign$ko <- rep(gsub(".rds","",gsub("kstest_pvadj_logfc_kovsctrl_sign_only_female_","", x)), length(kstest_pv_adj_kovsctrl_sign$circ))
    return(kstest_pv_adj_kovsctrl_sign)
}
```
