```r
pathways <- load_pathways("hsa")
```

```r
ko_sign_files <- list.files("/mnt/lustre/scratch/CBRA/research/projects/rrdd_mm/results/05_clasi_result_kstest/05_1_clasi_result_kstest_sign/")
```

```r
ko_kstest <-do.call(rbind,lapply(ko_sign_files, FUN=function(x) readRDS(paste("/mnt/lustre/scratch/CBRA/research/projects/rrdd_mm/results/05_clasi_result_kstest/05_1_clasi_result_kstest_sign/",x,sep=""))))
```

```r
ko_kstest_sign <- ko_kstest[ko_kstest$kspvadj < 0.05,]
```

```r
name_enfermerdad_symbol <- readRDS("/mnt/lustre/scratch/CBRA/research/projects/rrdd_mm/results/00_parser_orphanet/gen_enfermedad/name_enfermerdad_symbol.rds")
colnames(name_enfermerdad_symbol) <- c("name_enf","gene")
symbol_entreid <- AnnotationDbi::select(org.Hs.eg.db, keys=paste(name_enfermerdad_symbol$gene), columns=c("ENTREZID"),keytype="SYMBOL")

gen_enf <- unique(merge(name_enfermerdad_symbol, symbol_entreid, by.x="gene",by.y="SYMBOL"))
rd_gen_class <- readRDS("/mnt/lustre/scratch/CBRA/research/projects/rrdd_mm/results/00_parser_orphanet/rd_gen_class.rds")
```

```r
inf_rrdd <- merge(gen_enf, rd_gen_class, by.x=c("gene","name_enf"), by.y=c("V1","enfermedad"))
head(inf_rrdd)
```

```r
ko_kstest_sign$name_circ <- get_path_names(pathways, ko_kstest_sign$circ)
```

```r
ko_kstest_sign$path <- unlist(lapply(strsplit(ko_kstest_sign$name_circ, ":", fixed = T), FUN=function(x) x[1]))
```

```r
annot_ko_kstest_sign <- merge(inf_rrdd, ko_kstest_sign, by.x="ENTREZID", by.y="ko")
```
